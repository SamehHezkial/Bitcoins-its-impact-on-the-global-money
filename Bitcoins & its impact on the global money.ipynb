**-Step 7:Regression analysis:-**

```
# Set the independent dataset X_dat
# Change the categorical variable into dummies
X_dat = dat[['Regulation','Technology','Economy','Currency']]
X_dat
```

```
import statsmodels.api as sm
```

```
X_dat = sm.add_constant(X_dat)

X_dat
```

```
# Set the dependent variable (only one column)
y_dat = dat['Impact on USD']
y_dat
```

```
# Initial a model using OLS and fit the model. 
mod = sm.OLS(y_dat, X_dat)
res = mod.fit()
```

```
# Print the summary
print(res.summary())
```

**-Step 8: Notes on the regression analysis:-**
- The indepenent variables regression coefficients were having the same signs from what they had at the correlation matrix,except the **"Economy"** and that was beacuse of two reasons,first,the colinear relationship between the independent variables together and,second,the regression analysis is always taking into consideration the effect of all the independent variables together on the dependent one not one by one as the correlation does .. 
-The regression coefficients were too small values as well.
-At this point a question came into my mind which was,Can we rely on the **"Currency"** only as a predictor according to what we have recognized as long as it was the only predictor that kept it's sign & because economicswise there is always an inverse relationship between the availability of the currency & it's impact on the USD?,The real world cases say **NO** because not only one variable is taking control but more than one plus because of the second reason that we mentioned above at the first note.
- When we checked the P-values,all of them were greater than the significance level 0.1 which means that those variables (coefficientswise) aren't giving us a significant evidence about their relationship with the dependent variable.
-When we checked the linear regression report we found that the R squared value is too small which gave us a good  evidence about how fit is the model.
-Only one metric was left that we had to get in order to have a good judgement about the model which was the **"MSE"**.   

**- Calculating the MSE:**

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
trainX, testX, trainY, testY = train_test_split(dat[['Regulation','Technology','Economy','Currency']], dat["Impact on USD"], test_size=0.25)
```

```
# Initial a linear regression model, fit and predict
lr = LinearRegression()
lr.fit(trainX[['Regulation']], trainY)
lr_pred = lr.predict(testX[['Regulation']])
mean_squared_error(testY, lr_pred)
```

```
# Initial a linear regression model, fit and predict
lr = LinearRegression()
lr.fit(trainX[['Regulation','Technology']], trainY)
lr_pred = lr.predict(testX[['Regulation','Technology']])
mean_squared_error(testY, lr_pred)
```

```
# Initial a linear regression model, fit and predict
lr = LinearRegression()
lr.fit(trainX[['Regulation','Technology','Economy']], trainY)
lr_pred = lr.predict(testX[['Regulation','Technology','Economy']])
mean_squared_error(testY, lr_pred)
```

```
# Initial a linear regression model, fit and predict
lr = LinearRegression()
lr.fit(trainX[['Regulation','Technology','Economy','Currency']], trainY)
lr_pred = lr.predict(testX[['Regulation','Technology','Economy','Currency']])
mean_squared_error(testY, lr_pred)
```

- The value of the **MSE** decreased when we added more independent variables which was a good evidence about how fit is the model ..

**- Accuracy of the testing data set:**

```
lr.score(testX[['Regulation','Technology','Economy','Currency']],testY)
```

-Very low accuracy score.

**- Predicted values:**

```
pred = lr.predict(testX[['Regulation','Technology','Economy','Currency']])
print(pred)
```
